## Preprocessing

üëã ‚öí Now perform the following preprocessing steps and see how the number of unique words changes:


1.   Lower case all words in the text
2.   Remove punctuation markers and numbers (Hint: string.isalpha())
3.   Lemmatize all words in the text




```python
# Preprocess and then calculate the number of unique words in the text again
# 1
lemmatizer = WordNetLemmatizer()
words = word_tokenize(article.text)
lowercase_words = [word.lower() for word in words]

# 2
filtered_words = [word for word in lowercase_words if word.isalpha() and word not in string.punctuation]

# 3
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]


unique_words = set(filtered_words)
num_unique_words = len(unique_words)

print("Number of unique words after preprocessing: ", num_unique_words)
```

üëã ‚öí How many of these unique lemmas after preprocessing are in WordNet? Calculate the number by querying WordNet and check if the lemma is in WordNet or not.

```python
# Your code here

def is_in_wordnet(lemma):
    return len(wordnet.synsets(lemma)) > 0

num_in_wordnet = sum(1 for lemma in unique_words if is_in_wordnet(lemma))

print("Number of unique lemmas in WordNet: ", num_in_wordnet)
```

```python
import spacy
nlp_sm = spacy.load("en_core_web_sm")
```

```python
models = ["en_core_web_sm", "en_core_web_lg", "en_core_web_trf"]
nlp_models = {model: spacy.load(model) for model in models}

# Text from the article
text = article.text

# Perform NER and print the count of entities for each model
for model_name, nlp_model in nlp_models.items():
    doc = nlp_model(text)
    print(f"\nNamed Entity Recognition using {model_name} model:")

    # Count entities of each type
    entity_counts = {}
    for ent in doc.ents:
        entity_type = ent.label_
        entity_counts[entity_type] = entity_counts.get(entity_type, 0) + 1

    # Print the count of entities for each type
    for entity_type, count in entity_counts.items():
        print(f"{entity_type}: {count}")
```

# Home Exercise 2b: Evaluate analogy performance per relation type

üëã ‚öí Run through the Tutorial3.txt file from Tutorial3 and use your analogy function to evaluate how well the embeddings perform, however, this time calculate the accuracy for each relation type, e.g. `capital-common-countries`, separately. THe relationtypes are marked with `: `at the beginning of the line.  


```python
#Your code here

!wget https://github.com/dgromann/SemanticComputing/raw/master/tutorial6/word2vec_embeddings.bin
!wget https://raw.githubusercontent.com/dgromann/cl_intro/master/tutorials/Tutorial3.txt
!pip3 install gensim
```

```python
import gensim

# Load the word2vec model
model = gensim.models.KeyedVectors.load_word2vec_format("word2vec_embeddings.bin", binary=True)
```

```python
# analogy tasks
def read_analogy_tasks(file_path):
    tasks = []
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith(':'):
                relation_type = line[1:]
                current_task = {'relation_type': relation_type, 'examples': []}
            else:
                analogy_parts = line.split()
                current_task['examples'].append(analogy_parts)
                if len(current_task['examples']) == 3:
                    tasks.append(current_task)
    return tasks

# accuracy
def evaluate_accuracy(model, analogy_tasks):
    accuracy_by_relation = {}

    for task in analogy_tasks:
        relation_type = task['relation_type']
        correct_predictions = 0

        for example in task['examples']:
            a, b, c, expected_word = example
            predicted_word = model.most_similar(positive=[c, b], negative=[a], topn=1)[0][0]

            if predicted_word.lower() == expected_word.lower():
                correct_predictions += 1
                print(f"Analogy: {a}:{b} :: {c}:{predicted_word}  Expected: {expected_word}")

        accuracy = correct_predictions / len(task['examples'])
        accuracy_by_relation[relation_type] = accuracy

    return accuracy_by_relation

# read analogy tasks
file_path = "Tutorial3.txt"
analogy_tasks = read_analogy_tasks(file_path)

# evaluate accuracy
accuracy_by_relation = evaluate_accuracy(model, analogy_tasks)

# print the accuracy
for relation_type, accuracy in accuracy_by_relation.items():
    print(f"Accuracy for {relation_type}: {accuracy * 100:.2f}%")

```
